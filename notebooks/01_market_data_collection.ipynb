```python
import yfinance as yf
import pandas as pd
import numpy as np
```

Acciones Españolas

```python
tickers = {
    "Indra Sistemas": "IDR.MC",                   # Validado
    "Banco Santander": "SAN.MC",                  #
    "Iberdrola": "IBE.MC",
    "Inditex": "ITX.MC",
    "Telefónica": "TEF.MC",
    "Repsol": "REP.MC",
    "BBVA": "BBVA.MC",
    "Amadeus": "AMS.MC",
    "Ferrovial": "FER.MC",
    "Grifols": "GRF.MC"
}

# Rango de fechas
start_date = "2017-01-01"
end_date = "2025-04-30"

# Descargar datos mensuales
data = yf.download(
    tickers=list(tickers.values()),
    start=start_date,
    end=end_date,
    interval="1mo",
    group_by="ticker",
    auto_adjust=False
)

# DataFrame para consolidado final
consolidado = []

# Exportar a Excel con precio y rentabilidad logarítmica
with pd.ExcelWriter("data_esp_mensual_ESP_2017_2025.xlsx", date_format="DD-MM-YYYY") as writer:
    for name, symbol in tickers.items():
        df = data[symbol][["Adj Close"]].copy()
        df = df.reset_index()

        # Calcular rentabilidad logarítmica mensual
        df["Rent_Log"] = np.log(df["Adj Close"] / df["Adj Close"].shift(1))

        # Guardar para consolidado final
        df_consolidado = df[["Date", "Rent_Log"]].copy()
        df_consolidado["Accion"] = name
        consolidado.append(df_consolidado)

        # Guardar hoja individual
        df.to_excel(writer, sheet_name=name[:31], index=False)

df_final_esp = pd.concat(consolidado)
df_final_esp = df_final_esp.dropna() 
```

Consolidado de rentabilidades

```python
def crear_hoja_formato(df):
    df_pivot = df.pivot(index="Date", columns="Accion", values="Rent_Log")
    df_pivot = df_pivot.sort_index()
    return df_pivot

# Crear hojas por país
hoja_esp = crear_hoja_formato(df_final_esp)
hoja_fra = crear_hoja_formato(df_final_fra)
hoja_mex = crear_hoja_formato(df_final_mex)
hoja_deu = crear_hoja_formato(df_final_deu)
hoja_bra = crear_hoja_formato(df_final_bra)
hoja_usa = crear_hoja_formato(df_final_usa)
hoja_chl = crear_hoja_formato(df_final_chl)

# Exportar a Excel con hojas por país
with pd.ExcelWriter("rentabilidades_por_pais_2017_2025.xlsx", date_format="DD-MM-YYYY") as writer:
    hoja_esp.to_excel(writer, sheet_name="España")
    hoja_fra.to_excel(writer, sheet_name="Francia")
    hoja_mex.to_excel(writer, sheet_name="México")
    hoja_deu.to_excel(writer, sheet_name="Alemania")
    hoja_bra.to_excel(writer, sheet_name="Brasil")
    hoja_usa.to_excel(writer, sheet_name="EEUU")
    hoja_chl.to_excel(writer, sheet_name="Chile")

print("Archivo 'rentabilidades_por_pais_2017_2025.xlsx' generado correctamente.")
```
